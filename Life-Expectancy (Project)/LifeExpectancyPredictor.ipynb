__2. Finding Null Values__

```python
# Total missing values
dataset.isna().sum()
```

```python
sns.heatmap(data= dataset.isna(), cmap='viridis',cbar=False)
```

```python
#Features with Null/NaN
features_with_NaN = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]
print(features_with_NaN)
```

```python
#Percentage of Null value in a column
for feature in features_with_NaN:
    print(feature, np.round((dataset[feature].isnull().mean())*100,4),'% missing values')
```

_2.1 finding relationship between features with missing values and target feature_

```python
for feature in features_with_NaN:
    dataset_nan = dataset.copy()

    # index
    # 1 -> Null and 0 -> not null
    dataset_nan[feature] = np.where(dataset_nan[feature].isnull(),1 ,0)

    dataset_nan.groupby(feature)['Life expectancy '].median().plot.bar(cmap='viridis')
    plt.title(feature)
    plt.tight_layout()
    plt.show()
```
